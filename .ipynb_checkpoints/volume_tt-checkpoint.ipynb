Import the libraries that we will use at the beginning

### Returns the daily total volume of the three chains by product.

```python
chain_a = pd.read_csv("csv/chain_a.csv")
chain_b = pd.read_csv("csv/chain_b.csv")
chain_c = pd.read_csv("csv/chain_c.csv")
m_product = pd.read_csv("csv/m.product.csv")
```

Using pandas, I imported the csv sheets included in the same folder for analysis.

```python
chain_geral = pd.concat([chain_a, chain_b, chain_c], axis=0)
chain_geral.head()
```

As I exercise, the spreadsheets are from different directories, for that I need to concatenate them to have general access during the analysis

```python
conditionlist = [
    (chain_geral['store_id'] >= 1) & (chain_geral['store_id'] <= 3),
    (chain_geral['store_id'] >= 4) & (chain_geral['store_id'] <= 6),
    (chain_geral['store_id'] >= 7) & (chain_geral['store_id'] <= 9)]
choiselist = ['a','b','c']
chain_geral['chain'] = np.select(conditionlist, choiselist, default='Not Specified')
```

each chain group has a preset according to the exercise, to make monitoring easier, in this condition I fed a column to indicate each store belonging to each chain

```python
chain_geral
```

```python
geral_name = pd.merge(chain_geral, m_product, how="left", on="product_id")
```

```python
geral_name
```

```python
geral_name[["product_name", "volume"]].groupby("product_name").sum()
```

```python
end1 = geral_name.pivot_table(
    values="volume", index=["product_name","chain"], columns="Date", aggfunc="sum", margins=True
)
```
