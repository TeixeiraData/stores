```python
import pandas as pd
import numpy as np
```

Return month sales of 2 most expensive products in each product_category

```python
chain_a = pd.read_csv("csv/chain_a.csv")
chain_b = pd.read_csv("csv/chain_b.csv")
chain_c = pd.read_csv("csv/chain_c.csv")
m_product = pd.read_csv("csv/m.product.csv")
```

```python
m_product.head()
```

```python
m_productf=m_product[["product_id","product_category","price"]]
```

```python
m_productf["Rank"] = m_productf["price"].rank()
```

```python
m_productf.head()
```

```python
m_productf.sort_values("price", inplace = True)
```

```python
m_productf.head()
```

```python
chain_geral = pd.concat([chain_a, chain_b, chain_c], axis=0)
chain_geral.head()
```

```python
geral_name = pd.merge(chain_geral, m_product, how="left", on="product_id")
geral_name.head()
```

```python
geral_name.loc[geral_name["product_category"] == "Chocolte","product_category"] = "Chocolate"
```

```python
geral_name['Date'] = geral_name['Date'].astype('datetime64')
```

```python
geral_name.dtypes
```

```python
geral_name['month'] = geral_name['Date'].dt.month
geral_name['year'] = geral_name['Date'].dt.year
```

```python
geral_name.head()
```

```python
end2 = geral_name.pivot_table(
    values="volume", 
    index=["product_category" ,"price"], 
    columns=["year","month"], 
    aggfunc=np.sum
) 
end2
```

```python
end3 = end2.reset_index()
end3['price'] = pd.to_numeric(end3['price'])
end3
```

```python
end4 = end3.groupby('product_category')['price'].nlargest(2)
end3[end3['price'].isin(end4)].reset_index(drop = True)
```
